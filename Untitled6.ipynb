```python
import pandas as pd

# Load the dataset
movies = pd.read_csv('dataset.csv')

# Feature Selection
movies = movies[['id', 'title', 'overview', 'genre']]

# Fill any missing values with an empty string to avoid errors during concatenation
movies['overview'] = movies['overview'].fillna('')
movies['genre'] = movies['genre'].fillna('')

# Create 'tags' column by combining 'overview' and 'genre'
movies['tags'] = movies['overview'] + movies['genre']

# Drop the old columns
new_data = movies.drop(columns=['overview', 'genre'])

# Display the clean dataset
new_data.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Initialize CountVectorizer
cv = CountVectorizer(max_features=10000, stop_words='english')

# Transform tags into vectors
vector = cv.fit_transform(new_data['tags'].values.astype('U')).toarray()

# Calculate Cosine Similarity
similarity = cosine_similarity(vector)

print("Similarity matrix shape:", similarity.shape)
```

```python
def recommend(movies):
    # Find the index of the movie in the dataset
    index = new_data[new_data['title'] == movies].index[0]

    # Calculate distances and sort them
    distances = sorted(list(enumerate(similarity[index])), reverse=True, key=lambda x: x[1])

    # Print the top 5 most similar movies (skipping the first one, which is the movie itself)
    for i in distances[1:6]:
        print(new_data.iloc[i[0]].title)

# Test the function
print("Recommendations for 'Iron Man':")
recommend("Iron Man")
```

```python
import pickle

# Dump the dataframe and similarity matrix into pickle files
pickle.dump(new_data, open('movies_list.pkl', 'wb'))
pickle.dump(similarity, open('similarity.pkl', 'wb'))

# To verify it saved correctly, you can load it back
loaded_data = pickle.load(open('movies_list.pkl', 'rb'))
loaded_data.head()
```
